# Example usage

Here we will demonstrate how to use the `models` module to get the mean field approximation of the CAPSO model. 

## Imports

```python
import matplotlib.pyplot as plt
from pycapso.models import get_mean_field
```

## Simulate the mean field approximation

```python
prey_pred = pycapso.models.get_mean_field()
```

## Plotting the results

```python
plt.plot(prey_pred)
plt.xlabel("Time")
plt.ylabel("Density")
```
