# Example usage

Here we will demonstrate how to use the `get_mean_field()` function in the `models` module to get the mean field approximation of the CAPSO model. The function `plot_prey_pred_data()` in the `plots` module is used to plot the results.

## Imports

```python
import matplotlib.pyplot as plt
import seaborn as sns
from pycapso.models import get_mean_field
from pycapso.plots import plot_prey_pred_data
```

## Simulate the mean field approximation

```python
prey_pred = get_mean_field(num_iter = 50)
```

## Plotting the results

```python
sns.set_theme()
plot_prey_pred_data(prey_pred, prey_marker='.')
plt.show()
```
